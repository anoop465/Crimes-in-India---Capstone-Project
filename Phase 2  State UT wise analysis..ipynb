# Q. 2.1 Analysis of Literacy Rate vs Total Crimes.

```python
#Literacy Rate vs Total Crimes 
plt.figure(figsize=(10, 6))
sns.barplot(x='Literacy rate.', y='Crime Rate 2021', data=df, palette='Blues')
plt.title('Literacy Rate vs Total Crimes')
plt.xlabel('Literacy rate.')
plt.ylabel('Crime Rate 2021')
plt.xticks(rotation=90)
plt.grid(axis='y')
plt.show()


```

```python
# State-wise Crime Rate 2021
data_sorted = df.sort_values(by='Crime Rate 2021', ascending=False)

plt.figure(figsize=(12, 8))
bars = plt.bar(data_sorted['State / UT'], data_sorted['Crime Rate 2021'], color='Red')

for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width() / 2, yval, round(yval, 2), va='bottom', ha='center', fontsize=9)

plt.title('State-wise Crime Rate 2021')
plt.xlabel('State / UT')
plt.ylabel('Crime Rate 2021')
plt.xticks(rotation=90)  
plt.grid(axis='y')  
plt.tight_layout() 
plt.show()
```

```python
data_sorted = df.sort_values(by='Literacy rate.', ascending=False)

plt.figure(figsize=(12, 8))
bars = plt.bar(data_sorted['State / UT'], data_sorted['Literacy rate.'], color='Red')

for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width() / 2, yval, round(yval, 2), va='bottom', ha='center', fontsize=9)

plt.title('State-wise Crime Rate 2021')
plt.xlabel('State / UT')
plt.ylabel('Literacy rate.')
plt.xticks(rotation=90)  
plt.grid(axis='y')  
plt.tight_layout() 
plt.show()
```

# Q 2.2 Analysis of the type of crime vs each state vs Literacy rate.

```python
import pandas as pd

# Load the data from the CSV file
df = pd.read_excel("phase2data.xlsx")


# Display the column names to check for issues
print(df.columns)
```

```python

df.columns = df.columns.str.strip()

# Verify the columns
print(df.columns)

# Now proceed with melting the dataframe
df_melted = df.melt(id_vars=['State / UT', 'Literacy rate.'], 
                    var_name='Type of Crime', 
                    value_name='Crime Rate 2021')

# Display the first few rows to check the melted dataframe
print(df_melted.head())
```

# Q.2.3 Analysis of year-on-year total crime rate.

```python
crime_rates = df[['Crime Rate 2018', 'Crime Rate 2019', 'Crime Rate 2020', 'Crime Rate 2021']]

# Plotting
plt.figure(figsize=(12, 8))

# Stacked bar plot
crime_rates.plot(kind='bar', stacked=True, figsize=(12, 8))
plt.title('Year-on-Year Total Crime Rate')
plt.xlabel('State / UT')
plt.ylabel('Total Crime Rate')
plt.xticks(range(len(df['State / UT'])), df['State / UT'], rotation=90)
plt.legend(title='Year', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```

#  2.4 Analysis of area vs overall crime.

```python
# Plotting
plt.figure(figsize=(10, 6))
plt.scatter(df['Area km'], df['Total Crime 2021'], color='b', alpha=0.7)
plt.title('Area vs. Total Crime in 2021')
plt.xlabel('Area (km²)')
plt.ylabel('Total Crime (2021)')
plt.grid(True)
```

```python
# Box Plot of Total Crime
plt.figure(figsize=(8, 6))
sns.boxplot(data=df['Total Crime 2021'], color='lightgreen')
plt.title('Box Plot of Total Crime in 2021')
plt.xlabel('Total Crime in 2021')
plt.show()
```

```python
# Histogram of Crime Distribution
plt.figure(figsize=(12, 6))
plt.hist(df['Total Crime 2021'], bins=15, color='skyblue', edgecolor='black')
plt.title('Distribution of Total Crime in 2021')
plt.xlabel('Total Crime in 2021')
plt.ylabel('Number of States/UTs')
plt.grid(True)
plt.show()
```

```python
#  Scatter Plot of Area vs. Crime Rate in 2021
df['Crime Rate 2021'] = df['Total Crime 2021'] / df['Area km']
plt.figure(figsize=(10, 6))
plt.scatter(df['Area km'], df['Crime Rate 2021'], color='red', alpha=0.7)
plt.title('Area vs. Crime Rate in 2021')
plt.xlabel('Area (km²)')
plt.ylabel('Crime Rate in 2021 (Crimes per km²)')
plt.grid(True)
```

```python
#Bar Chart of Top 10 States/UTs by Total Crime in 2021
top_10_crime = df.nlargest(10, 'Total Crime 2021')
plt.figure(figsize=(14, 8))
sns.barplot(x='Total Crime 2021', y='State / UT', data=top_10_crime, palette='viridis')
plt.title('Top 10 States/UTs by Total Crime in 2021')
plt.xlabel('Total Crime in 2021')
plt.ylabel('State / UT')
plt.show()
```

#  2.5 Analysis of Population vs overall Crime.

```python
# Plotting
plt.figure(figsize=(12, 8))
plt.scatter(df['Population'], df['Total Crime 2021'], color='b', alpha=0.7)
plt.title('Population vs Overall Crime Rate')
plt.xlabel('Population')
plt.ylabel('Total Crime (2021)')
plt.grid(True)
plt.xticks(rotation=90)
plt.tight_layout()
```

```python
# Adding a Regression Line to the Scatter Plot
slope, intercept = np.polyfit(df['Population'], df['Total Crime 2021'], 1)
plt.plot(df['Population'], slope * df['Population'] + intercept, color='red', linewidth=2)
```
